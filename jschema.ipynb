```python
    def map_container(callable):
        @functools.wraps(callable)
        def logic(object, *args, **kwargs):
            if isinstance(object, (list, dict)): return object.map(callable, *args, **kwargs)
            return object.pipe(callable, *args, **kwargs)
        return logic
```
